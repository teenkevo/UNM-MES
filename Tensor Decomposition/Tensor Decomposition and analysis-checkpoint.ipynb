## RESCAL Factorization of the Linked Building Data Model (Alternating Least Squares)

```python
# IMPORTING DEPENDENCIES
import rdflib
from rdflib import Graph, URIRef, Literal, BNode
from rdflib.namespace import RDF, RDFS, OWL
from rdflib.extras.external_graph_libs import rdflib_to_networkx_graph, rdflib_to_networkx_multidigraph
import networkx as nx
import matplotlib.pyplot as plt
import logging
import itertools
import numpy as np
from numpy.linalg import norm
from numpy.random import shuffle
from scipy.io.matlab import loadmat
from scipy.sparse import lil_matrix
from sktensor import dtensor, rescal, cp_als
from sklearn.metrics import precision_recall_curve, auc, roc_auc_score
```

```python
# SOME UTILITY FUNCTIONS

# Function for loading the RDF graph and creating a 3-way tensor
def rdf2tensor(rdf_graph, graph_serialization):
    # create rdflib Graph object
    g = Graph()
    # load graph
    # g.parse(rdf_graph, format = graph_serialization)
    g.parse(rdf_graph, format = graph_serialization)
    
    # SUBJECT set
    subject_set = set()
    # datatype and object PROPERTY sets
    datatype_property_set = set()
    object_property_set = set()
    # literal and non_literal OBJECT sets
    literal_object_set = set()
    non_literal_object_set = set()

    # iterate through graph to populate the above lists
    for sub, pred, obj in g:
        subject_set.add(sub)
        # check for LITERAL OBJECTS and the DATATYPE-PROPERTIES connected to them
        if type(obj) == rdflib.term.Literal:
            literal_object_set.add(obj)
            datatype_property_set.add(pred)
        # check for NON-LITERAL OBJECTS objects and the OBJECT-PROPERTIES connected to them
        else:
            non_literal_object_set.add(obj)
            object_property_set.add(pred)
    
    # [Nodes for OBJECT PROPERTY Graph]- A combination of the SUBJECT SET with the NON-LITERAL OBJECT SET
    nodes4_opg = subject_set|non_literal_object_set
    # conversion to a list to allow index accessibility
    nodes4_opg = list(subject_set|non_literal_object_set)
    # CREATE a 3-WAY TENSOR for the OBJECT-PROPERTIES populated with zeros
    opg_tensor = np.zeros((
                           len(object_property_set), 
                           len(nodes4_opg), 
                           len(nodes4_opg)
                         ))
    # UPDATE entties of the opg_tensor with 1 where relation exists, leave 0 otherwise 
    for key, obj_pred in enumerate(object_property_set):
        k = key
        for s, p, o in g:
            if obj_pred == p:
                opg_tensor[k][nodes4_opg.index(s)][nodes4_opg.index(o)] = 1

    # [Nodes for DATATYPE PROPERTY Graph]- A combination of the SUBJECT SET with the NON-LITERAL OBJECT SET
    nodes4_dpg = subject_set|literal_object_set
    # conversion to a list to allow index accessibility 
    nodes4_dpg = list(subject_set|literal_object_set)
    # create a 3-WAY TENSOR for the DATATYPE-PREDICATES populated with zeros
    dpg_tensor = np.zeros((
                           len(datatype_property_set), 
                           len(nodes4_dpg), 
                           len(nodes4_dpg)
                         ))
    # UPDATE entties of dpg_tensor with 1 where relation exists, leave 0 otherwise 
    for key, dp_pred in enumerate(datatype_property_set):
        j = key
        for s, p, o in g:
            if dp_pred == p:
                dpg_tensor[j][nodes4_dpg.index(s)][nodes4_dpg.index(o)] = 1
    return opg_tensor, dpg_tensor

def normalize_predictions(prediction, num_entities, num_relations):
    for a in range(num_entities):
        for b in range(num_entities):
            nrm = norm(prediction[:num_relations, a, b])
            if nrm != 0:
                # round values for faster computation of AUC-PR
                prediction[:num_relations, a, b] = np.round_(prediction[:num_relations, a, b] / nrm, decimals=3)
    return prediction

# Function for factorizing the tensorized RDF Linked Building Data graph and reconstruction of the tensor from the factors
def rescal_fact(train_tensor, n_dim, num_entities, num_relations):
    # calculate the factorization
    entity_embedding, R, _, _, = rescal.als(train_tensor, n_dim, init='nvecs', conv=1e-3, lambda_A=1, lambda_R=1)
    n = entity_embedding.shape[0]
    reconstructed_tensor = np.zeros((len(R), n, n))
    for k in range(len(R)):
        reconstructed_tensor[k, :, :] = np.dot(entity_embedding, np.dot(R[k], entity_embedding.T))
    reconstructed_tensor_non_normalized = np.round(reconstructed_tensor, 3)
    # normalization the reconstruction
    reconstructed_tensor_normalized = normalize_predictions(reconstructed_tensor, num_entities, num_relations)

    return entity_embedding, reconstructed_tensor_non_normalized, reconstructed_tensor_normalized

# Function for loading the linked building data 3-way tensor from which a test/train split done 
def load_data(input_tensor, train_fraction=0.7):
    # un comment the two lines below if loading data from a MATLAB file & change the input tensor argument to filename
    # mat = loadmat(filename)
    # K = np.array(mat['Rs'], np.float32)
    num_entities, num_relations = input_tensor.shape[1], input_tensor.shape[0]
    # construct array for rescal
    T = [lil_matrix(input_tensor[i, :, :]) for i in range(num_relations)]
    
    # Train Test Split
    triples = num_entities * num_entities * num_relations
    IDX = list(range(triples))
    shuffle(IDX)
    train = int(train_fraction*len(IDX))
    idx_test = IDX[train:]

    train_tensor = [Ti.copy() for Ti in T]
    mask_idx = np.unravel_index(idx_test, (num_relations, num_entities, num_entities))
    # set values to be predicted to zero
    for i in range(len(mask_idx[0])):
        train_tensor[mask_idx[0][i]][mask_idx[1][i], mask_idx[2][i]] = 0
    return input_tensor, train_tensor, mask_idx, num_entities, num_relations

# Function similar to above but tailored for hyper parameter tuning (Use the one above for training after choosing optimal parameters with this one)
def load_train_val_test(input_tensor, train_fraction=0.6, val_fraction=0.2):
    # un comment the two lines below if loading data from a MATLAB file & change the input tensor argument to filename
    # mat = loadmat(filename)
    # K = np.array(mat['Rs'], np.float32)
    num_entities, num_relations = input_tensor.shape[1], input_tensor.shape[0]
    # construct array for rescal
    T = [lil_matrix(input_tensor[i, :, :]) for i in range(num_relations)]
    
    # Train Test Split
    triples = num_entities * num_entities * num_relations
    IDX = list(range(triples))
    shuffle(IDX)
    train = int(train_fraction*len(IDX))
    val = int(val_fraction*len(IDX))
    idx_val = IDX[train:train+val]
    idx_test = IDX[train+val:]

    train_tensor = [Ti.copy() for Ti in T]
    mask_idx = np.unravel_index(idx_test+idx_val, (num_relations, num_entities, num_entities))
    val_idx = np.unravel_index(idx_val, (num_relations, num_entities, num_entities))
    test_idx = np.unravel_index(idx_test, (num_relations, num_entities, num_entities))
    # set values to be predicted to zero
    for i in range(len(mask_idx[0])):
        train_tensor[mask_idx[0][i]][mask_idx[1][i], mask_idx[2][i]] = 0
    return input_tensor, train_tensor, val_idx, test_idx, num_entities, num_relations
```

```python
# HYPER_PARAMETER TUNING
n_dim = 100
rdf_graph = 'dht22.ttl'
graph_serialization = 'ttl'

opg_tensor, dpg_tensor = rdf2tensor(rdf_graph, graph_serialization)
input_tensor = opg_tensor

# Mask Test and Validation Triples
input_tensor, train_tensor, val_idx, test_idx, num_entities, num_relations = load_train_val_test(input_tensor, train_fraction=0.6, val_fraction=0.2)
var_list = [0.001, 0.1, 1., 10., 100.]
best_roc = 0
# Train on training set and evaluate on validation set
for (var_x, var_e, var_r) in itertools.product(var_list, repeat=3):
    A, R, f, itr = rescal.als(train_tensor, n_dim, lambda_A=var_x, lambda_R=var_e, lambda_V=var_r)
    n = A.shape[0]
    reconstructed_tensor = np.zeros((len(R), n, n))
    for k in range(len(R)):
        reconstructed_tensor[k, :, :] = np.dot(A, np.dot(R[k], A.T))
    reconstructed_tensor = normalize_predictions(reconstructed_tensor, num_entities, num_relations)
    score = roc_auc_score(input_tensor[val_idx], reconstructed_tensor[val_idx])

    print('var_x:{0:3.3f}, var_e:{1:3.3f}, var_r:{2:3.3f}, AUC-ROC:{3:.3f}'.format(var_x, var_e, var_r, score))
    if score > best_roc:
        best_vars = (var_x, var_e, var_r)
        best_roc = score
        
lambda_a, lambda_r, lambda_v = best_vars
print(best_vars, best_roc)
# Use optimum parameters on Validation Set and Test on Hold-Out Set
lambda_a, lambda_r, lambda_v = best_vars
A, R, f, itr = rescal.als(train_tensor, n_dim, lambda_A=var_x, lambda_R=var_e, lambda_V=var_r)
n = A.shape[0]
reconstructed_tensor = np.zeros((len(R), n, n))
for k in range(len(R)):
    reconstructed_tensor[k, :, :] = np.dot(A, np.dot(R[k], A.T))
reconstructed_tensor = normalize_predictions(reconstructed_tensor, num_entities, num_relations)
score = roc_auc_score(input_tensor[test_idx], reconstructed_tensor[test_idx])
print('AUC On Test Set with Optimum Parameters\n{}'.format(score))
```

```python
# Using OPTIMUM parameters to train the RESCAL model of the Linked Building Data RDF tensor
n_dim = 100
rdf_graph = 'dht22.ttl'
graph_serialization = 'ttl'

opg_tensor, dpg_tensor = rdf2tensor(rdf_graph, graph_serialization)
input_tensor = opg_tensor

input_tensor, train_tensor, target_idx, num_entities, num_relations = load_data(input_tensor, train_fraction=0.7)
# Train Rescal
entity_embedding, reconstructed_tensor_non_normalized, reconstructed_tensor_normalized = rescal_fact(train_tensor, n_dim, num_entities, num_relations)
prec, recall, _ = precision_recall_curve(input_tensor[target_idx], reconstructed_tensor_normalized[target_idx])
#entities = mat['names']
print('AUC\n{}'.format(roc_auc_score(input_tensor[target_idx], reconstructed_tensor_normalized[target_idx])))

plt.plot(recall, prec, marker='.', label='Logistic')
```
